```python
# Imports
import shutil
import warnings
import subprocess
import numpy as np
from Bio import Phylo
from Bio import SeqIO
from Bio import AlignIO
from Bio.Seq import Seq
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
from Bio.Phylo.TreeConstruction import DistanceCalculator
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
def compute_distance_from_alignment(file_path: str):
    alignment = AlignIO.read(file_path, "fasta")
    calculator = DistanceCalculator('blastn')
    matrix = calculator.get_distance(alignment)
    sequence_names = [record.id for record in alignment]
    return matrix, sequence_names

def plot_distance_matrix(matrix, sequence_names, figure_width=8, figure_height=6):
    plt.figure(figsize=(figure_width, figure_height))
    heatmap = plt.imshow(matrix, cmap='coolwarm', aspect='auto')
    plt.colorbar(heatmap, label="Distance")
    plt.xticks(ticks=range(len(sequence_names)), labels=sequence_names, rotation=90)
    plt.yticks(ticks=range(len(sequence_names)), labels=sequence_names)
    for i in range(len(sequence_names)):
        for j in range(len(sequence_names)):
            plt.text(j, i, f"{matrix[i][j]:.2f}", ha='center', va='center', color='black')
    plt.xlabel("Sequence")
    plt.ylabel("Sequence")
    plt.title("Distance Matrix")
    plt.tight_layout()
    plt.show()

def find_optimal_clusters(distance_matrix):
    best_silhouette = -1
    best_num_clusters = 0
    best_cluster_labels = None
    for num_clusters in range(2, len(distance_matrix)):
        clusterer = AgglomerativeClustering(n_clusters=num_clusters, affinity='precomputed', linkage='complete')
        cluster_labels = clusterer.fit_predict(distance_matrix)
        silhouette_avg = silhouette_score(distance_matrix, cluster_labels, metric='precomputed')
        
        print(f"For n_clusters = {num_clusters}, average silhouette score is: {silhouette_avg}")
        
        if silhouette_avg > best_silhouette:
            best_silhouette = silhouette_avg
            best_num_clusters = num_clusters
            best_cluster_labels = cluster_labels

    return best_num_clusters, best_silhouette, best_cluster_labels
```

```python
def plot_colored_dendrogram(file_path, distance_matrix, num_clusters):
    linkage_matrix = linkage(distance_matrix, method='complete')
    cluster_labels = fcluster(linkage_matrix, num_clusters, criterion='maxclust')

    # Convert RGB array to hex for string comparison
    cluster_palette = [plt.matplotlib.colors.rgb2hex(c) for c in plt.cm.rainbow(np.linspace(0, 1, num_clusters))]

    cluster_colors = {i: cluster_palette[i] for i in range(num_clusters)}

    leaf_colors = {i: cluster_colors[cluster_labels[i] - 1] for i in range(len(cluster_labels))}

    def get_link_color(link_index):
        num_items = len(distance_matrix)
        
        # Return the cluster color for individual data points
        if link_index < num_items:
            return leaf_colors[link_index]
        
        # Get the children of the current link
        left, right = int(linkage_matrix[link_index - num_items, 0]), int(linkage_matrix[link_index - num_items, 1])
        
        # If the child is an individual data point
        if left < num_items:
            left_color = leaf_colors[left]
        # If the child is another link
        else:
            left_color = get_link_color(left)
        
        if right < num_items:
            right_color = leaf_colors[right]
        else:
            right_color = get_link_color(right)

        # If both children have the same color, return that color for the current link
        if left_color == right_color:
            return left_color
        # Otherwise return black (or any other default color)
        else:
            return "black"

    plt.figure(figsize=(10, 7))
    labels = [seq.id for seq in AlignIO.read(file_path, "fasta")]

    # Tracez le dendrogramme avec les labels récupérés
    plt.figure(figsize=(10, 7))
    dendrogram(linkage_matrix, labels=labels, link_color_func=get_link_color, orientation='right')
    plt.title('Dendrogram with Colored Branches')
    plt.xticks(rotation=90) 
    plt.tight_layout()
    plt.xlim(left=-0.0025)
    plt.show()
```
