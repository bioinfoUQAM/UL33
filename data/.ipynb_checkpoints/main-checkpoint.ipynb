```python
# Imports
import os
import time
import pysam
import shutil
import subprocess
from pyfaidx import Fasta
import numpy as np
from Bio import Phylo
from Bio import SeqIO
import matplotlib.pyplot as plt
from Bio.SeqRecord import SeqRecord
from Bio import AlignIO
import matplotlib.pyplot as plt
from Bio.Phylo.TreeConstruction import DistanceCalculator
from Bio.Phylo.TreeConstruction import DistanceTreeConstructor
```

```python
def generate_strandcount(sample_name):
    # Define the command as a single string
    command = "java -cp lib/htsjdk-unspecified-SNAPSHOT.jar:lib/picocli-4.1.2.jar:lib/pal-1.5.1.jar:lib/cache2k-all-1.0.2.Final.jar:lib/commons-math3-3.6.1.jar:jar/MakeReadCount.jar makereadcount.MakeReadCount data/" + sample_name + "/" + sample_name +"27.M1.W10.07/27.M1.W10.07.bam"
    # Execute the command and capture the output
    result = subprocess.run(command, shell=True, capture_output=True, text=True, check=True)
    # Display the output
    print(result.stdout)
```

```python
sample_name = "30.B1.W11.08"
generate_strandcount(sample_name)
```

```python
def align_sequences(sample_name):
    # Align the consensus sequences using Muscle
    command = "muscle -in tree/" + sample_name + ".fasta -out tree/" + sample_name + "_aligned.fasta"
    subprocess.run(command, check=True, shell=True)
```

```python
def generate_tree(sample_name):
    # Generate a phylogenetic tree using FastTree
    with open("tree/" + sample_name + "_aligned.fasta", "r") as input_file, open("tree/" + sample_name + "_consensus_sequences.tree", "w") as output_file:
        subprocess.run(["fasttree", "-nt", "-boot", "1000"], stdin=input_file, stdout=output_file, check=True)

    # Draw the phylogenetic tree
    tree = Phylo.read("tree/" + sample_name + "_consensus_sequences.tree", "newick")
    fig = plt.figure(figsize=(12, 5)) 
    ax = fig.add_subplot(1, 1, 1)
    Phylo.draw(tree, axes=ax, label_func=lambda x: str(x)[:])
    plt.show()
```

```python
def compute_distance_matrixe(sample_name):
    # Define the path to the FASTA file containing the aligned sequences
    file_path = "tree/" + sample_name + "_aligned.fasta"
    # Open the FASTA file using Biopython's AlignIO
    alignment = AlignIO.read(file_path, "fasta")
    # Use Jaccard distance to calculate the similarity matrix
    calculator = DistanceCalculator('identity')
    matrix = calculator.get_distance(alignment)
    # Create a heatmap of the similarity matrix
    plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
    heatmap = plt.imshow(matrix, cmap='coolwarm', aspect='auto')
    # Add colorbar
    plt.colorbar(heatmap, label="Distance")
    # Set the sequence names as tick labels on the axes
    sequence_names = [record.id for record in alignment]
    plt.xticks(ticks=range(len(sequence_names)), labels=sequence_names, rotation=90)
    plt.yticks(ticks=range(len(sequence_names)), labels=sequence_names)
    # Display the similarity values within the heatmap cells
    for i in range(len(sequence_names)):
        for j in range(len(sequence_names)):
            plt.text(j, i, f"{matrix[i][j]:.2f}", ha='center', va='center', color='black')

    # Set the axis labels and title
    plt.xlabel("Sequence")
    plt.ylabel("Sequence")
    plt.title("Distance Matrix")
    # Adjust the layout to prevent label cutoff
    plt.tight_layout()
    # Display the heatmap
    plt.show()
```

```python
# Family 4
sample_name = "04.B1.W14.01_04.M1.W09.02"
align_sequences(sample_name)
generate_tree(sample_name)
compute_distance_matrixe(sample_name)
```

```python
# Family 5
sample_name = "05.B1.W14.04_05.M1.W08.03"
align_sequences(sample_name)
generate_tree(sample_name)
compute_distance_matrixe(sample_name)
```
